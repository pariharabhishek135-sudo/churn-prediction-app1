#### Bivariate Analysis


1.Scatterplot


2.Countplot


3.Boxplot

```python
# CreditScore vs Balance
sns.scatterplot(x="CreditScore", y="Balance", data=df)
plt.show()
```

#### Multivariate Analysis


Heatmap

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")

plt.show()
```
